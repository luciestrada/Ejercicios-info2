```python
import matplotlib.pyplot as plt
import numpy as np

x= np.linspace(0,2* np.pi,200) #200 puntos
#ver los vectores
print(x)#tamaño de x
y= np.sin(x)

#axisis, los ejes
fig, ax = plt.subplots()
ax.plot(x,y,label="Linea")
#Nombres de los ejes
ax.set_xlabel("Eje x")
ax.set_ylabel("Eje y")
ax.set_title("Gráfica de la función seno")
ax.legend()
plt.show()
```

**Crear dos axis, en una sola funcion: Es crear dos figuras en la misma grafica**



```python
import matplotlib.pyplot as plt
import numpy as np

x1= np.linspace(0,3* np.pi,200) 
y1= np.sin(x1)

x2= np.linspace(0,2* np.pi,200) 
y2= np.cos(x2)

fig, (ax1, ax2 )= plt.subplots(2,1,figsize=[10,10]) # El 2,1 es como el orden. 2 filas , y una columna. y si pone(1,2)= ..... figsize= tamaño de la figura

ax1.plot(x1,y1,'o',label="Grafica seno",color="brown", linewidth=25)
ax2.plot(x2,y2,'v',label="Grafica coseno",color="brown", linewidth=25 )

ax1.set_xlabel("Eje x")
ax1.set_ylabel("Eje y")
ax1.set_title("Gráfica de la función seno")
ax1.legend()
ax1.text(6,0.50,"Aqui estoy yoo")
ax1.grid() #PONER CUADRICULA

ax2.set_xlabel("Eje x")
ax2.set_ylabel("Eje y")
ax2.set_title("Gráfica de la función seno")
ax2.legend()
ax2.grid()

#Poner flecha
ax2.annotate('aquí estas tú', xy=(2,-0.25), xytext=(3,0.5),arrowprops=dict(facecolor= "blue", shrink=0.05))

plt.show()
```
