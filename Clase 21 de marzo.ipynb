## Interpolaci칩n

Hallar puntos, que ayude a encontrar un punto en la mitad, y necesitamos uno final y uno inicial.

```python
from scipy.interpolate import interp1d
import numpy as np  
import matplotlib.pyplot as plt # para hacer la grafica

datos_x = np.arange(0, 0.1, 0.005) # rango entre 0 y 0.1 , y 0.005 es el pase
A, k, theta = 10.0, 33.3, np.pi/5.0  # Asignando variables
y_real = A*np.sin(2*np.pi*k*datos_x + theta) # A * el seno de todo eso
datos_y = y_real + 2*np.random.randn(len(datos_x)) # y_real + un vector que tenga las cantidades de datos_x

interpolador_lineal = interp1d(datos_x, datos_y) # interpolacion de los graficos de los daots de X y Y


x_inter = np.linspace(0.012, 0.09, 500) # Va de 0.012 hasta 0.09 y tiene 500 cosas que las separa
y_inter_l = interpolador_lineal(x_inter)


plt.plot(datos_x, datos_y, 'ok', label="Datos") # ok = puntos
plt.plot(x_inter, y_inter_l, 'r', label="Interpolaci칩n lineal") # r = linea
plt.title("Interpolaci칩n") # titulo
plt.xlim(-0.003, 0.102) # los limites de los numeros, desde -0.003
plt.ylim(-14, 15)
plt.legend()
```
