```python
import numpy as np

# Crear una matriz de ceros de tipo entero 3x4
matriz_ceros = np.zeros((3, 4), dtype=int)
print("Matriz de ceros:")
print(matriz_ceros)

# Crear una matriz de ceros de tipo entero 3x4 excepto la primera fila que será uno
matriz_unos_primera_fila = np.zeros((3, 4), dtype=int)
matriz_unos_primera_fila[0] = 1
print("\nMatriz de ceros con la primera fila de unos:")
print(matriz_unos_primera_fila)

# Crear una matriz de ceros de tipo entero 3x4 excepto la última fila que será el rango entre 5 y 8
matriz_rango_5_8_ultima_fila = np.zeros((3, 4), dtype=int)
matriz_rango_5_8_ultima_fila[-1] = np.arange(5, 9)
print("\nMatriz de ceros con la última fila como el rango entre 5 y 8:")
print(matriz_rango_5_8_ultima_fila)

# Crear un vector de 10 elementos, siendo los índices impares unos y los índices pares dos
vector = np.ones(10, dtype=int)
vector[::2] = 2
print("Vector de 10 elementos con índices impares unos y pares dos:")
print(vector)

# Crear un "tablero de ajedrez", con unos en las casillas negras y ceros en las blancas
tablero = np.zeros((8, 8), dtype=int)
tablero[1::2, ::2] = 1  # Filas impares, columnas pares
tablero[::2, 1::2] = 1  # Filas pares, columnas impares
print("\nTablero de ajedrez:")
print(tablero)

# Crear una matriz aleatoria 5x5
matriz_aleatoria = np.random.rand(5, 5)
print("Matriz aleatoria 5x5:")
print(matriz_aleatoria)

# Hallar los valores mínimo y máximo
minimo = np.min(matriz_aleatoria)
maximo = np.max(matriz_aleatoria)
print("\nValor mínimo:", minimo)
print("Valor máximo:", maximo)


```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

n = np.arange(1,99) #Intervalos de valores n
y_real = np.sin(np.pi * 0.12 * n) # Construcción de la señal y[n] utilizando la fórmula sin(π0.12m)

plt.figure(figsize=(10, 6))
plt.plot(n, y_real, 'ok', label="Datos",color='purple') #Ok genera los puntos y label es la linea

interpolador_lineal = interp1d(n, y_real, kind='linear') #no es necesario el kind-linear

x_inter = np.linspace (1, 50, 40) #son los intervalos del eje x del interpolador: 1=>donde empieza, 50=>donde termina, 40=> los puntos que une
y_inter = interpolador_lineal(x_inter)

plt.plot(x_inter, y_inter, '-', label='y [n] Interpolado', color='lightblue')

plt.title("señal y[n] = sin(π0.12n)") #nombre de la gráfica
plt.xlabel('n') #nombre del eje x
plt.ylabel('amplitud') #nombre del eje y
plt.legend()
```

```python
class MatrixOperations:

    def __init__(self, seed=42):
        # Inicializar la semilla del generador de números aleatorios
        np.random.seed (seed)

    def randomMatrices (self):
#Crear matrices aleatorias con dimensiones 3x3 y 5x2
        matrix_3x3 = np.random.randint(1, 10, size=(3, 3))
        matrix_5x2 = np.random.randint(1, 10, size=(5, 2))
        return matrix_3x3, matrix_5x2

    def suma(self, A, B):
    # Sumar matrices A y B
        if A. shape == B.shape:
            return A + B
        else:
            raise ValueError("Las dimensiones de las matrices deben ser iguales para sumar")

    def resta(self, A, B):
# Restar matrices A y B
        if A. shape == B. shape:
            return A - B
        else: 
            raise ValueError("Las dimensiones de las matrices deben ser iguales para restar")

    def multiplicacion (self, A, B):
#Multiplicar matrices A y B 
        if A.shape[1] == B.shape[0]:
            return np.dot(A, B)
        else:
            raise ValueError("El número de columnas de A debe ser igual al número de filas de B para multiplicar")
        

    def chessBoard(self):
    # Crear un tablero de ajedrez de 8x8
        board = np.zeros((8, 8), dtype=str)
        board[::2, ::2] = '#'
        board[1::2, 1::2] = '#'
        board[::2, 1::2] = '.'
        board[1::2, ::2] = '.'
        return board

    def matrizCruz(self, size):
    # Crear una matriz cuadrada con una cruz de unos en el centro
        cruz = np.zeros((size, size))
        centro = size // 2
        cruz [centro, :] = 1
        cruz[:, centro] = 1
        return cruz

    def matrizDiagonal (self, size):
    # Crear una matriz cuadrada nxn con diagonal aleatoria
        matriz = np.zeros((size, size))
        diagonal_values = np.random.randint(1, 11, size)
        np.fill_diagonal (matriz, diagonal_values)
        return matriz

    # Crear una instancia de la clase MatrixOperations
    operations = MatrixOperations()

    # Probar los métodos de la clase

    # Matrices aleatorias
    matriz_3x3, matriz_5x2 = operations.randomMatrices()
    print("Matriz 3x3 aleatoria:")
    print(matriz_3x3)
    print("\nMatriz 5x2 aleatoria:")
    print(matriz_5x2)

    # Tablero de ajedrez
chess_board = operations.chessBoard()
print("\nTablero de ajedrez:")
print(chess_board)

# Matriz con cruz
matriz_cruz = operations.matrizCruz(5)
print("\nMatriz con cruz:")
print(matriz_cruz)

# Matriz con diagonal aleatoria
matriz_diagonal = operations.matrizDiagonal(4)
print("\nMatriz con diagonal aleatoria:")
print(matriz_diagonal)
```
