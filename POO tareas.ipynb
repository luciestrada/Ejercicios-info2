```python
class Tarea:
    def __init__(self, descripcion, prioridad):
        self.descripcion = descripcion
        self.prioridad = prioridad
        self.completada = False

    def marcar_completada(self):
        self.completada = True

    def editar_descripcion(self, nueva_descripcion):
        self.descripcion = nueva_descripcion

    def editar_prioridad(self, nueva_prioridad):
        self.prioridad = nueva_prioridad

class ListaTareas:
    def __init__(self):
        self.tareas = []

    def agregar_tarea(self, tarea):
        self.tareas.append(tarea)

    def mostrar_tareas_pendientes(self):
        print("Tareas Pendientes:")
        for i, tarea in enumerate(self.tareas):
            if not tarea.completada:
                print(f"{i}. Descripción: {tarea.descripcion}, Prioridad: {tarea.prioridad}")

    def mostrar_tareas_completadas(self):
        print("Tareas Completadas:")
        for i, tarea in enumerate(self.tareas):
            if tarea.completada:
                print(f"{i}. Descripción: {tarea.descripcion}, Prioridad: {tarea.prioridad}")

def mostrar_menu():
    print("\nMenú:")
    print("1. Agregar nueva tarea")
    print("2. Marcar tarea como completada")
    print("3. Editar descripción de tarea")
    print("4. Editar prioridad de tarea")
    print("5. Mostrar tareas pendientes")
    print("6. Mostrar tareas completadas")
    print("7. Salir")

if __name__ == "__main__":
    lista_tareas = ListaTareas()

    while True:
        mostrar_menu()
        opcion = input("Seleccione una opción: ")

        if opcion == "1":
            descripcion = input("Ingrese la descripción de la nueva tarea: ")
            prioridad = input("Ingrese la prioridad de la nueva tarea (Alta/Media/Baja): ").capitalize()
            nueva_tarea = Tarea(descripcion, prioridad)
            lista_tareas.agregar_tarea(nueva_tarea)
            print("Tarea agregada correctamente.")

        elif opcion == "2":
            lista_tareas.mostrar_tareas_pendientes()
            tarea_index = int(input("Ingrese el índice de la tarea que desea marcar como completada: "))
            if tarea_index < len(lista_tareas.tareas):
                lista_tareas.tareas[tarea_index].marcar_completada()
                print("Tarea marcada como completada.")
            else:
                print("Índice de tarea no válido.")

        elif opcion == "3":
            lista_tareas.mostrar_tareas_pendientes()
            tarea_index
```
