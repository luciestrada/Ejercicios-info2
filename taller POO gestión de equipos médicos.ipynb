1. SISTEMA DE GESTIÓN DE SIGNOS VITALES
Se propone diseñar un sistema para la gestión de signos vitales de pacientes en un servicio de hospitalización. El sistema debe gestionar médicos y pacientes. Los pacientes tienen información de nombre, cédula, temperatura, frecuencia cardiaca, presión. Los médicos tienen información de nombre, cédula, registromédico y la lista de los pacientes que tiene a cargo. El médico puede verificar el estado de un paciente o ingresar un nuevo paciente. Para verificar el estado de unpaciente se ingresa su número de cédula y el sistema deberá retornar una cadena con el siguiente formato:
Nombre; Temperatura; Frecuencia cardiaca; Presión, en el orden indicado y usando los ‘;’. En caso de que el
número de identificación no exista se debe retornar la cadena: Paciente no existe. Para ingresar un paciente se solicita su identificación, en caso de que exista se retorna la cadena: Paciente
existe. En el otro caso se ingresa los datos restantes, y una vez creado se agrega a la lista de pacientes del
sistema.

```python
class Paciente:  # Corrección del nombre de la clase a singular (Paciente en lugar de Pacientes)
    def __init__(self, nombre, cedula, temp, fcar, presion):  # Corrección de la definición del método __init__
        self.nombre = nombre
        self.cedula = cedula
        self.temp = temp
        self.fcar = fcar
        self.presion = presion

class Medicos:
    def __init__(self, nombre , cedula , registromedico ):
        self.nombre=nombre
        self.cedula=cedula
        self.registromedico=registromedico
        self.lista_pacientes=[]


    def ingresa_paciente(self,paciente):#metodo de agregar
        for p in self.lista_pacientes: #p recorre la lista, y mira si está
            if p.cedula == paciente.cedula: #si la p.decula es igual a la cedula del paciente -->existe
              return "Paciente existe"

        self.lista_pacientes.append(paciente)
        return "Paciente ingresado correctamente"


    def verificar(self, cedula_paciente):
        for paciente in self.lista_pacientes:
            if paciente.cedula == cedula_paciente:
                return f"{paciente.nombre}; {paciente.temp}; {paciente.fcar} ; {presion} "
        
        return print("Paciente no existe")

class SistemaGestionSignosVitales:#Si toca verificar los datos del medico
    def __init__(self):
        self.medicos = []

    def agregar_medico(self, medico):
        self.medicos.append(medico)

    def buscar_medico_por_cedula(self, cedula_medico):
        for medico in self.medicos:
            if medico.cedula == cedula_medico:
                return medico
        return None


sistema = SistemaGestionSignosVitales() 
medico1 = Medicos("Dr. Juan Perez", "123456", "M1234")
medico2 = Medicos("Dr. María López", "654321", "M5678")
sistema.agregar_medico(medico1)
sistema.agregar_medico(medico2)


while True:
    print("1. ingresar Paciente")
    print("2. verificar")
    print ("3. Salir")
    menu=input("Ingresar opción:")

    if menu == "1":
        id_medico= input("ingrese la cedula del medico:")
        medico = sistema.buscar_medico_por_cedula(id_medico)#busca si está la cedula
        if medico:
            nombre = input("Ingrese el nombre del paciente: ")
            cedula = input("Ingrese la cédula del paciente: ")
            temperatura = float(input("Ingrese la temperatura del paciente: "))
            frecuencia_cardiaca = int(input("Ingrese la frecuencia cardíaca del paciente: "))
            presion = input("Ingrese la presión del paciente: ")

            paciente = Paciente(nombre, cedula, temperatura, frecuencia_cardiaca, presion)#clase paciente
            print(medico.ingresa_paciente(paciente))
        else:
            print("Médico no encontrado.")

    elif menu == "2":
        cedula_paciente = input("Ingrese la cédula del paciente: ")
        medico_cedula = input("Ingrese la cédula del médico: ")
        medico = sistema.buscar_medico_por_cedula(medico_cedula)
        if medico:
            print(medico.verificar(cedula_paciente))
        else:
            print("Médico no encontrado.")

    

        
```
