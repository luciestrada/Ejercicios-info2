```python
class Animal:
    def __init__(self, nombre, especie, edad, dueño):
        self.nombre = nombre
        self.especie = especie
        self.edad = edad
        self.dueño = dueño

class Consulta:
    def __init__(self, animal, fecha, motivo, diagnostico):
        self.animal = animal
        self.fecha = fecha
        self.motivo = motivo
        self.diagnostico = diagnostico

class ClinicaVeterinaria:
    def __init__(self):
        self.pacientes = []
        self.consultas = []

    def agregar_animal(self, animal):
        if animal not in self.pacientes:
            self.pacientes.append(animal)
        else:
            print("El animal ya está registrado en la clínica.")

    def programar_consulta(self, animal, fecha, motivo, diagnostico):
        consulta = Consulta(animal, fecha, motivo, diagnostico)
        self.consultas.append(consulta)
        print("Consulta programada para", animal.nombre, "el", fecha)

    def mostrar_consultas(self):
        print("Consultas realizadas:")
        for consulta in self.consultas:
            print("Animal:", consulta.animal.nombre)
            print("Fecha:", consulta.fecha)
            print("Motivo:", consulta.motivo)
            print("Diagnóstico:", consulta.diagnostico)
            print()

if __name__ == "__main__":
    clinica = ClinicaVeterinaria()

    while True:
        print("Bienvenido a la Clínica Veterinaria")
        print("1. Agregar nuevo animal")
        print("2. Programar consulta")
        print("3. Mostrar consultas realizadas")
        print("4. Salir")

        opcion = input("Ingrese el número de la opción que desea realizar: ")

        if opcion == "1":
            nombre = input("Ingrese el nombre del animal: ")
            especie = input("Ingrese la especie del animal: ")
            edad = int(input("Ingrese la edad del animal: "))
            dueño = input("Ingrese el nombre del dueño del animal: ")
            animal = Animal(nombre, especie, edad, dueño)
            clinica.agregar_animal(animal)
        elif opcion == "2":
            nombre_animal = input("Ingrese el nombre del animal: ")
            animal = None
            for paciente in clinica.pacientes:
                if paciente.nombre == nombre_animal:
                    animal = paciente
                    break
            if animal:
                fecha = input("Ingrese la fecha de la consulta (YYYY-MM-DD): ")
                motivo = input("Ingrese el motivo de la consulta: ")
                diagnostico = input("Ingrese el diagnóstico de la consulta: ")
                clinica.programar_consulta(animal, fecha, motivo, diagnostico)
            else:
                print("El animal no está registrado en la clínica.")
        elif opcion == "3":
            clinica.mostrar_consultas()
        elif opcion == "4":
            print("Gracias por usar nuestro sistema. ¡Hasta luego!")
            break
        else:
            print("Opción no válida. Por favor, seleccione una opción válida.")
```
